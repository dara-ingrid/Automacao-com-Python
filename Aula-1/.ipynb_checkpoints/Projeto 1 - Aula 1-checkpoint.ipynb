# Automação de Sistemas e Processos com Python

### Desafio:

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadorias da empresa.
O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos.

E-mail do seu chefe: para o nosso exercício, coloque um e-mail seu como sendo o e-mail do seu chefe<br>
Link de acesso ao sistema da empresa: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
#!pip install pyautogui
```

```python
import pyautogui
import time
link = "https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema"
#Passo a Passo

pyautogui.PAUSE = 2

#Passo 1: Acessar o sistema 
pyautogui.hotkey('ctrl', 't')  #abrir nova aba
pyautogui.write(link)
pyautogui.press("enter")
time.sleep(3)

#Passo 2: Preencher o login e senha
#print(pyautogui.position())  #capturar a posição do mouse
pyautogui.click(x=755, y=567)
pyautogui.write("meu_login")
pyautogui.click(x=898, y=683)
pyautogui.write("minha_senha")
pyautogui.click(x=925, y=783) #Clica no botão Acessar
time.sleep(3)

#Passo 3: Fazer download da base de dados

pyautogui.click(x=591, y=442)  #Selecionar o arquivo
pyautogui.click(x=1641, y=291)  #Clicar nos 3 pontinhos
time.sleep(5)
pyautogui.click(x=1508, y=845)  #Selecionar fazer download
time.sleep(8)
pyautogui.click(x=1085, y=662) #clicar em Salvar
time.sleep(10)
pyautogui.click(x=1891, y=964) #fecha a tela de downloads
```

```python
#Passo 4: Calcular os indicadores
import pandas as pd

tabela = pd.read_csv(r"C:\Users\dara-\Downloads\Compras.csv", sep=";")
display(tabela)
total_gasto = tabela["ValorFinal"].sum()
quantidade = tabela["Quantidade"].sum()
preco_medio = total_gasto / quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)
```

```python
import pyperclip

#Passo 5: Enviar um e-mail para o meu chefe

#Entrar no site do email
pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://mail.google.com/mail/u/0/#inbox")
pyautogui.press("enter")

time.sleep(10)

#Clicar em Escrever
#print(pyautogui.position())
pyautogui.click(x=130, y=286)

#Preencher as informações do email
pyautogui.write("dara.ingrid313@gmail.com") #digita o e-mail
pyautogui.press("tab")  #sair do campo de destinatário

#print(pyautogui.position())
pyautogui.press("tab") #Avança para o campo de Assunto
pyperclip.copy("Relatório de Vendas - Teste Intensivão Python")
pyautogui.hotkey("ctrl", "v")

pyautogui.press("tab") #mudar para o corpo do e-mail

texto = f"""
Presados, 
segue o Relatório de compras

Total Gasto: R$ {total_gasto:,.2f}
Quantidade de Produtos: {quantidade:,}
Preço Médio: R$ {preco_medio:,.2f}

Qualquer dúvida, estou à disposição

At.te, Dara
"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

#Enviar
pyautogui.hotkey("ctrl", "enter")
```
